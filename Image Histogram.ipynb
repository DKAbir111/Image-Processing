# Import the libraries

```python
import cv2
import pandas
import numpy as np
import matplotlib.pyplot as plt 
```

# Calculate Histogram of gray image

# Histogram of dark image

```python
# find frequency of pixels in range 0-255

img1=cv2.imread("C:/Users/HP/ImageData/dark.png",0) # 0 value for coverting BGR to Gray image
plt.imshow(img1,cmap='gray')
plt.show()


# cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
histr1 = cv2.calcHist([img1],[0],None,[256],[0,256])


# show the plotting graph of an image
plt.plot(histr1)
plt.show()
```

```python
img3=cv2.imread("C:/Users/HP/ImageData/highcontrast.png", 0)
plt.imshow(img3,cmap='gray')
plt.show()
histr3 = cv2.calcHist([img3],[0],None,[256],[0,256])
plt.plot(histr3)
plt.xlabel("Gray level")
plt.ylabel("No of pixels")
plt.show()
```

# Histogram Equalization

```python
img = cv2.imread('C:/Users/HP/ImageData/low1.jpg', 0)
  
# creating a Histograms Equalization of a image
equ = cv2.equalizeHist(img)
  
# stacking images side-by-side
res = np.hstack((img, equ))
  
# show image input vs output

plt.imshow(img)
plt.show()
plt.imshow(equ)
plt.show()
plt.imshow(res)
plt.show()
```

# Color Image Histogram

```python
image=cv2.imread("C:/Users/HP/ImageData/dark.png")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
b,g,r=cv2.split(img)
```

```python
histrb = cv2.calcHist([b],[0],None,[256],[0,256])
plt.plot(histrb)

histrg = cv2.calcHist([g],[0],None,[256],[0,256])
plt.plot(histrg)

histrr = cv2.calcHist([r],[0],None,[256],[0,256])
plt.plot(histrr)
plt.show()
```

# Color Image Histogram Equalization

```python
image=cv2.imread("C:/Users/HP/ImageData/dark.png")
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv) 

h,s,v=cv2.split(img_hsv)
plt.axis("off")
```

```python
v= cv2.equalizeHist(v)
merged_hsv=cv2.merge((h,s,v))
bgr_enhanced=cv2.cvtColor(merged_hsv,cv2.COLOR_HSV2RGB)
plt.axis("off")
plt.imshow(bgr_enhanced)
plt.show()
```

```python
#Histogram of color image
r,g,b=cv2.split(image)
histrb = cv2.calcHist([r],[0],None,[256],[0,256])
plt.plot(histrb)

histrg = cv2.calcHist([g],[0],None,[256],[0,256])
plt.plot(histrg)

histrr = cv2.calcHist([b],[0],None,[256],[0,256])
plt.plot(histrr)
plt.show()
```

```python
#Equalized Histogram of color image
r,g,b=cv2.split(bgr_enhanced)
histrb = cv2.calcHist([r],[0],None,[256],[0,256])
plt.plot(histrb)

histrg = cv2.calcHist([g],[0],None,[256],[0,256])
plt.plot(histrg)

histrr = cv2.calcHist([b],[0],None,[256],[0,256])
plt.plot(histrr)
plt.show()
```

# Contrast Limited Adaptive Histogram E

```python
img = cv2.imread('C:/Users/HP/ImageData/hist1.PNG', 0)
plt.imshow(img)
plt.axis("off")
plt.show()

clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8, 8))
equ = clahe.apply(img)
plt.imshow(equ)
plt.axis("off")
plt.show()
```

```python
img = cv2.imread('C:/Users/HP/ImageData/clahe.PNG', 0)
plt.imshow(img)
plt.axis("off")
plt.show()


equ = cv2.equalizeHist(img)
plt.imshow(equ)
plt.axis("off")
plt.show()

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
equ1 = clahe.apply(img)
plt.imshow(equ1)
plt.axis("off")
plt.show()
```

```python
hist1 = cv2.calcHist([img],[0],None,[256],[0,255])
plt.plot(hist1)
plt.show()
hist2 = cv2.calcHist([equ],[0],None,[256],[0,255])
plt.plot(hist2)
plt.show()

hist3 = cv2.calcHist([equ1],[0],None,[256],[0,255])
plt.plot(hist3)
plt.show()
```
