```python
import torch
import torchvision
import kornia
import cv2
import pandas
import numpy as np
import matplotlib.pyplot as plt 
```

# Bilateral filter

```python
#read an image
image=cv2.imread("b.jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#image to tensor
tensor_image=kornia.utils.image_to_tensor(img, keepdim=False).float()

#bilateral filter
output = kornia.filters.bilateral_blur(tensor_image, (25, 25), 350, (11.5, 11.5))

#tensor to image
img_blur = kornia.tensor_to_image(output.byte())

#subplotting
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img_blur)
```

# Blur using max pooling

```python
#read an image
image=cv2.imread("b.jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#image to tensor
tensor_image=kornia.utils.image_to_tensor(img, keepdim=False).float()

#blur pool 2D
output = kornia.filters.blur_pool2d(tensor_image, 11)

#tensor to image
img_blur = kornia.tensor_to_image(output.byte())

#subplotting
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img_blur)
```

# Guided filter

```python
#Read an RGB image
image1=cv2.imread("Test1.png")
img1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image2=cv2.imread("Tiger.jpg")
img2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

#Convert an image to tensor
tensor_image1=kornia.utils.image_to_tensor(img1, keepdim=False).float()
tensor_image2=kornia.utils.image_to_tensor(img2, keepdim=False).float()

#Guided filter
output = kornia.filters.guided_blur(tensor_image1,tensor_image2, 3,0.1)

#Convert a tensor to image
output_img1= kornia.tensor_to_image(output.byte())

#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,3,1)
ax1.imshow(img1)
ax2 = fig.add_subplot(1,3,2)
ax2.imshow(img2)
ax3 = fig.add_subplot(1,3,3)
ax3.imshow(output_img1)
```

# Median filter

```python
#read an image
image=cv2.imread("sp.png")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#image to tensor
tensor_image=kornia.utils.image_to_tensor(img, keepdim=False).float()

#Median filter
output = kornia.filters.median_blur(tensor_image,(15, 15))

#tensor to image
img_blur = kornia.tensor_to_image(output.byte())

#subplotting
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img_blur)
```

# Motion Blur
