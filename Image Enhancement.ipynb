# Import libraries

```python
import torch
import torchvision
import kornia
import cv2
import pandas
import numpy as np
import matplotlib.pyplot as plt 
```

# Calculate the weighted sum of two Tensors

```python
#Read an RGB image
image1=cv2.imread("Test1.png")
img1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image2=cv2.imread("Tiger.jpg")
img2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

#Convert an image to tensor
tensor_image1=kornia.utils.image_to_tensor(img1, keepdim=False).float()
tensor_image2=kornia.utils.image_to_tensor(img2, keepdim=False).float()

#Calculate the weighted sum of two Tensors
output= kornia.enhance.add_weighted(tensor_image1, 0.5,tensor_image2, 0.5, 1.0)


#Convert a tensor to image
output_img1= kornia.tensor_to_image(output.byte())


#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,3,1)
ax1.imshow(img1)
ax2 = fig.add_subplot(1,3,2)
ax2.imshow(img2)
ax3 = fig.add_subplot(1,3,3)
ax3.imshow(output_img1)
```

# posterize: Reduce the number of bits for each color channel

# Normalization

```python
#Read an RGB image
image1=cv2.imread("equalization.png")
img1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

#Convert an image to tensor
tensor_image1=kornia.utils.image_to_tensor(img1).float() / 255.

#equalization
output= kornia.enhance.normalize_min_max(tensor_image1, min_val=-1., max_val=1.)

#Convert a tensor to image
output_img1= kornia.tensor_to_image(output)

#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img1)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(output_img1)
```
